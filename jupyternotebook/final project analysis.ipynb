```python
import pandas as pd
import numpy as np

# Load the dataset
df=pd.read_csv(r"C:\Users\john8\Desktop\python\projet dataset\country_year_avg_temp.csv")

# find the minimum year for each country
min_years = df.groupby('Country')['Year'].min()
print(min_years)

# find the minimum year for each country
min_years = df.groupby('Country')['Year'].min()
# count the number of countries with minimum year less than 1850
count = (min_years > 1850).sum()
print(count)
```

```python
new_dfs = []

for country, min_year in min_years.items():
    if min_year > 1850:
        # create a DataFrame with the missing years
        years = pd.DataFrame({'Year': range(1850, min_year)})
        years['Country'] = country
        years['Avg_Temperature'] = np.nan

        # filter the original DataFrame for the current country and years >= min_year
        subset = df[(df['Country'] == country) & (df['Year'] >= min_year)]

        # combine the new DataFrame and the filtered subset
        new_df = pd.concat([years, subset])

        # add the new DataFrame to the list of modified DataFrames
        new_dfs.append(new_df)
    else:
        # if min_year <= 1850, no modification is needed, so just add the original DataFrame to the list
        new_dfs.append(df[df['Country'] == country])

# concatenate all modified DataFrames into a single DataFrame
new_df = pd.concat(new_dfs, ignore_index=True)
new_df
```

```python
import pandas as pd
import numpy as np

# Group the data by country
grouped = new_df.groupby('Country')

# Initialize an empty list to store the DataFrames with filled years
filled_years_dataframes = []

# Iterate over the groups
for country, group in grouped:
    all_years = range(1850, 2014)  # Specify the range of years, including 2013
    existing_years = group['Year'].values
    missing_years = list(set(all_years) - set(existing_years))

    # Create a DataFrame with the missing years
    missing_years_df = pd.DataFrame({'Year': missing_years})
    missing_years_df['Country'] = country
    missing_years_df['Avg_Temperature'] = np.nan

    # Concatenate the group data and the missing years DataFrame
    filled_group = pd.concat([group, missing_years_df])
    filled_years_dataframes.append(filled_group)

# Combine all filled groups into a single DataFrame named 'new_df'
new_df = pd.concat(filled_years_dataframes)

# Sort the data by country and year
new_df = new_df.sort_values(['Country', 'Year']).reset_index(drop=True)

print(new_df)
```

```python
# Filter the dataset by the year 1850 and later
new_df_after_1850 = new_df[new_df['Year'] >= 1850]

# Group the filtered data by the country column
grouped_by_country = new_df_after_1850.groupby('Country')

# Print the first 5 rows of each group
for name, group in grouped_by_country:
    print(name)
    print(group.head(5))
```

```python
# Fill missing values in Avg_Temperature column using backward filling
new_df['Avg_Temperature'].fillna(method='bfill', inplace=True)

# Filter the data for years greater than or equal to 1850
new_df_after_1850 = new_df[new_df['Year'] >= 1850]

# Convert the Year column to integer data type
new_df_after_1850['Year'] = new_df_after_1850['Year'].astype(int)

# Group the filtered data by the country column
grouped_by_country = new_df_after_1850.groupby('Country')

# Print the first 5 rows of each group
for name, group in grouped_by_country:
    print(name)
    print(group.head(5))

# Save the cleaned data to a CSV file
new_df_after_1850.to_csv('cleaned_country_year_avg_temp.csv', index=False)
```
