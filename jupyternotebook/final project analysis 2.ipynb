```python
import pandas as pd

# Read the dataset from the CSV file
df = pd.read_csv(r'C:\Users\john8\Desktop\python\projet dataset\cleaned_country_year_avg_temp.csv')

# Filter the DataFrame for years 1850 and 2013
df_1850 = df[df['Year'] == 1850]
df_2013 = df[df['Year'] == 2013]

# Merge the two DataFrames on the 'Country' column
merged_df = pd.merge(df_1850, df_2013, on='Country', suffixes=('_1850', '_2013'))

# Calculate the temperature difference
merged_df['Temp_Difference 1850-2013'] = merged_df['Avg_Temperature_2013'] - merged_df['Avg_Temperature_1850']

# Create a new DataFrame with only the desired columns
result_df = merged_df[['Country', 'Temp_Difference 1850-2013']]

# Display the result DataFrame
print(result_df)
```

```python
# Sort the result DataFrame by 'Temperature_Difference' in descending order
result_df = result_df.sort_values(by='Temp_Difference 1850-2013', ascending=False)
# Display the sorted result DataFrame
print(result_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the result_df DataFrame already
result_df = result_df.sort_values(by='Temp_Difference 1850-2013', ascending=False)

# Select the top 5 countries
top_five = result_df.head(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_five['Country'], top_five['Temp_Difference 1850-2013'],color='red')

# Customize the plot
plt.title('Top 10 Countries with the Highest Temperature Differences (1850-2013)')
plt.xlabel('Country')
plt.ylabel('Temperature Difference (°C)')
# Rotate x-axis labels to prevent overlapping
plt.xticks(rotation=45)
# Show the plot
plt.show()
```

```python
import pandas as pd

df2 = pd.read_csv(r'C:\Users\john8\Desktop\python\projet dataset\cleaned_us_states_avg_temp.csv')

# Filter the DataFrame for years 1850 and 2013
df2_1850 = df2[df2['Year'] == 1850]
df2_2013 = df2[df2['Year'] == 2013]

# Merge the two DataFrames on the 'State' column
merged_df2 = pd.merge(df2_1850, df2_2013, on='State', suffixes=('_1850', '_2013'))

# Calculate the temperature difference
merged_df2['Temp_Difference 1850-2013'] = merged_df2['Avg_Temperature_2013'] - merged_df2['Avg_Temperature_1850']

# Create a new DataFrame with only the desired columns
result_df2 = merged_df2[['State', 'Temp_Difference 1850-2013']]

# Display the result DataFrame
print(result_df2)
```

```python
# Sort the result DataFrame by 'Temperature_Difference' in descending order
result_df2 = result_df2.sort_values(by='Temp_Difference 1850-2013', ascending=False)
# Display the sorted result DataFrame
print(result_df2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Create the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(result_df2['State'], result_df2['Temp_Difference 1850-2013'])

# Customize the plot
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Temperature Difference (1850-2013)')
plt.title('Temperature Difference between 1850 and 2013 for US States')

# Display the plot
plt.show()
```
