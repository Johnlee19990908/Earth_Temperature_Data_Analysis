```python
#bycountry dataset missing value distribution
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load the dataset and preprocess it
dataframe = pd.read_csv(r'C:\Users\john8\Desktop\python\projet dataset\GlobalLandTemperaturesByCountry.csv')
dataframe.rename(columns={"dt": "Date"}, inplace=True)
dataframe = dataframe.set_index('Date')
dataframe.drop_duplicates(inplace=True)
dataframe.rename(columns={"AverageTemperature": "Avg_Temperature"}, inplace=True)
dataframe.rename(columns={"AverageTemperatureUncertainty": "Avg_Temperature_Uncertainty"}, inplace=True)

# Drop rows with all missing values
dataframe.dropna(how='all', inplace=True)

# convert 'Date' index to datetime
dataframe.index = pd.to_datetime(dataframe.index)

# calculate the percentage of missing values
missing_percentage = dataframe.isnull().sum() / len(dataframe) * 100

# create a pivot table of missing values by year
missing_year = dataframe.groupby(dataframe.index.year).apply(lambda x: x.isnull().sum())

# create a heatmap
sns.heatmap(missing_year, cmap='Reds')

# add title and labels
plt.title('Distribution of Missing Values in Temperature Data')
plt.xlabel('Month')
plt.ylabel('Year')

# show the plot
plt.show()
```
